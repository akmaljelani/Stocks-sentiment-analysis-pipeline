## 4.2 Strip out unwanted URLs

```python
## 2 rules. Have https in the string & exclude policies, maps. Using splitting function & lookup
import re
```

```python
exclude_list = ['maps','policies', 'preferences', 'accounts', 'support']
```

```python
## res is regular expression library
## split fro https:// to '&'' value
## val will hold the value
## append the res result into val
## convert val array into a set. THen convert into a list. To avoid any duplicate

def strip_unwanted_urls(urls, exclude_list):
    val = []
    for url in urls: 
        if 'https://' in url and not any(exclude_word in url for exclude_word in exclude_list):
            res = re.findall(r'(https?://\S+)', url)[0].split('&')[0]
            val.append(res)
    return list(set(val))
```

```python
## stripping unwanted urls from raw urls, then passing exclude list

strip_unwanted_urls(raw_urls['TSLA'], exclude_list)
```

```python
cleaned_urls = {ticker:strip_unwanted_urls(raw_urls[ticker], exclude_list) for ticker in monitored_tickers}
cleaned_urls
```

## 4.3 Search & Scrape Cleaned URLs

1. similar to scraping one article above.
2. Searching 'p' for paragraphs
3. Looping through each paragraphs
4. summarize 350 words limitation
5. Storing in an empty array [] ARTICLES

```python
def scrape_and_process(URLs):
    ARTICLES = []
    for url in URLs: 
        r = requests.get(url)
        soup = BeautifulSoup(r.text, 'html.parser')
        paragraphs = soup.find_all('p')
        text = [paragraph.text for paragraph in paragraphs]
        words = ' '.join(text).split(' ')[:350]
        ARTICLE = ' '.join(words)
        ARTICLES.append(ARTICLE)
    return ARTICLES
```

```python
## list comprehension & storing arrays in a dictionary

articles = {ticker:scrape_and_process(cleaned_urls[ticker]) for ticker in monitored_tickers}
articles
```

# 4.4 Summarize all articles

1. new summarize functions to loop one ticker at a time.
2. Looping through each article
3. Summarization runs on GPU. May take time

```python
def summarize(articles):
    summaries = []
    for article in articles:
        input_ids = tokenizer.encode(article, return_tensors='pt')
        output = model.generate(input_ids, max_length=55, num_beams=5, early_stopping=True)
        summary = tokenizer.decode(output[0], skip_special_tokens=True)
        summaries.append(summary)
    return summaries
```

```python
summaries = {ticker:summarize(articles[ticker]) for ticker in monitored_tickers}
summaries
```

# 6. Exploring Results into CSV

1. passing summaries, scores, urls to a blank array (output)
2. Looping through ticker
3. Looping through individual summary per ticker
4. Using counter to get corresponding sentiment score & url for the summary
5. len(summaries) means running through how many summaries per ticker. In this case 10 summaries for 'TSLA'
6. Output_this = tying summaries back to ticker code, looping through summaries['TSLA'][0],[1],[2]...
7. Appending output_this to output empty array


```python
def create_output_array(summaries, scores, urls):
    output = []
    for ticker in monitored_tickers:
        for counter in range(len(summaries[ticker])):
            output_this = [
                ticker,
                summaries[ticker][counter],
                scores[ticker][counter]['label'],
                scores[ticker][counter]['score'],
                urls[ticker][counter]
            ]
            output.append(output_this)
    return output
```

```python
final_output = create_output_array(summaries, scores, cleaned_urls)
final_output
```

```python
## adding header

final_output.insert(0, ['Ticker', 'Summary', 'Label', 'Confidence', 'URL'])
```
